```python
from classification import train_test_split, feature_generator
df.rename(columns={"preprocessed_posts" : "preprocessed_text"}, inplace=True)
X_train, X_test, y_train, y_test, y_train_enc, y_test_enc = train_test_split(df=df, test_size=0.3, rs=42069)


X_train_vec, none, vectorizer = feature_generator(X_train, X_test, y_train, 2, "tfidf")
```

```python
from sklearn.linear_model import LogisticRegression

classifier = LogisticRegression(solver = "lbfgs", max_iter=3000)
classifier.fit(X_train_vec, y_train)
```

```python
y_pred = classifier.predict(vectorizer.transform(X_test["preprocessed_text"]))
```

```python


from sklearn.metrics import classification_report, confusion_matrix, accuracy_score, f1_score

print(confusion_matrix(y_test,y_pred))
print(classification_report(y_test,y_pred))
print(accuracy_score(y_test, y_pred))
print(f1_score(y_test, y_pred, average='weighted'))
```

```python
import joblib
import pickle

joblib.dump(classifier, "LR_Models" + os.sep + "LR_Classifier_16_tfidf_unbalanced_noGrid.pkl")
```

Unbalanced With GridSearchCV

```python
import numpy as np
from sklearn.model_selection import GridSearchCV
# parameter grid
parameters = {
    'penalty' : ['l1','l2'], 
    'C'       : np.logspace(-3,3,7),
    'solver'  : ['newton-cg', 'lbfgs', 'liblinear'],
}
```

```python
logreg = LogisticRegression()
clf = GridSearchCV(logreg,                    # model
                   param_grid = parameters,   # hyperparameters
                   scoring='f1_micro',        # metric for scoring
                   cv=10)

            
```

```python
clf.fit(X_train_vec, y_train)
```

```python
print("Tuned Hyperparameters :", clf.best_params_)
print("F1 :",clf.best_score_)
```

```python
logreg = LogisticRegression(C = 1.0, 
                            penalty = 'l1', 
                            solver = 'liblinear')
logreg.fit(X_train_vec,y_train)

y_pred = logreg.predict(vectorizer.transform(X_test["preprocessed_text"]))

print(confusion_matrix(y_test,y_pred))
print(classification_report(y_test,y_pred))
print(accuracy_score(y_test, y_pred))
print(f1_score(y_test, y_pred, average='weighted'))
```

```python
joblib.dump(logreg, "LR_Models" + os.sep + "LR_Classifier_16_tfidf_unbalanced_Grid.pkl")
```

Balanced no Grid

```python
from imblearn.over_sampling import RandomOverSampler
# balance data
sampler = RandomOverSampler()
X_train_balanced, Y_train_balanced = sampler.fit_resample(X_train, y_train)
```

```python
X_train_vec_balanced = vectorizer.transform(X_train_balanced["preprocessed_text"])
```

```python
logreg = LogisticRegression()
logreg.fit(X_train_vec_balanced, Y_train_balanced)

y_pred = logreg.predict(vectorizer.transform(X_test["preprocessed_text"]))

print(confusion_matrix(y_test,y_pred))
print(classification_report(y_test,y_pred))
print(accuracy_score(y_test, y_pred))
print(f1_score(y_test, y_pred, average='weighted'))
```
