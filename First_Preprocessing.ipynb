```python
import pandas as pd

df = pd.read_csv("data\mbti_1.csv")
```

```python
def convert_posts_to_list(post: str) -> list:

    list = post.split("|||")
    list = post.lstrip()
    return list
```

```python
def convert_posts_to_one_text(post: str) -> str:
    post = post.replace("|||", "")
    post = post.replace("'", "")
    post = post.replace('"', "")

    
    
```

```python
df["posts"] = df["posts"].apply(convert_posts_to_list)
```

```python
def remove_quotation_from_orig_scrape(string: str) -> str:

    if len(string) > 0:
        
        if string[0] == "\"" or string[0] == "\'":
            string = string[1:]

        if string[-1] == "\"" or string[-1] == "\'":
            string = string[:-1]
    
    return string
```

```python
df2["posts"] = df2["posts"].apply(remove_quotation_from_orig_scrape)
```

```python
import validators
```

```python
validators.url("https://stackoverflow.com/questions/7160737/how-to-validate-a-url-in-python-malformed-or-not")
```

```python
def post_is_only_url(post: str) -> bool:

    is_only_url = validators.url(post)

    if is_only_url == True:
        return is_only_url
    else:
        return False
```

```python
df2.head()
```

```python
df2.reset_index()
```

```python
df2['only_url'] = df2.apply(lambda x: post_is_only_url(x["posts"]), axis=1)
```

```python
df2.reset_index()
```

```python
df_only_url_removed = df2.loc[df2["only_url"] == False]
```

```python
df_only_url_removed = df_only_url_removed.drop('only_url' , axis=1)
```

```python
df_only_url_removed.reset_index()
```

```python
def remove_quotation(post: str):
    post = post.replace('"', '')
    post = post.replace("'", "")

    return post
```

```python
df_only_url_removed["posts"] = df_only_url_removed["posts"].apply(remove_quotation)
```
